## 必要なPythonの文法知識：if文

`if`文は条件分岐が必要な処理を書くときに利用する。以下のように書くことができる

```python
if 条件式:
    条件式の結果が真（True）のときの処理
else:
    条件式の結果が偽（False）のときの処理
```

条件式を定義するためには，論理演算子を利用する。論理演算子はたくさんあるが，以下の3つを紹介する。

- `==` 等しい
- `<` より小さい
- `>` より大きい

````python
x = 1
x == 1 # 1と等しいかどうか
````

````python
x > 0 # 0より大きいかどうか
````

````python
x < 0 # 0より小さいかどうか
````

例えば，if 文を使うことで絶対値を求める関数を定義することができる。

$$
|x| = \begin{cases}
    x & if x \geq 0 \\
    -x & if x < 0
  \end{cases}
$$

````python
# 絶対値を求める関数 絶対値は absolute value
def absol(x):
    # xが0より小さい時，- をかけた値を返す
    if x < 0:
        return -x
    # xが0より小さくない（つまり，0以上の）とき，そのまま値を返す
    else:
        return x
````

````python
absol(3)
````
