## 補遺

$x|z$, $y|z$ は何なのか，「zの影響を取り除く」とはどういうことなのか。ここでは$x|z$ を取り上げて，上記の式からこれらの解釈を試みる（$y|z$ についても同様である）。



第2項（括弧`()`の部分）について，一旦，相関係数の部分（$r_{xz}$）を無視すると，標準化された z に x の標準偏差 $SD_x$ をかけたのち，x の平均を足している。これは，z の目盛りを x の目盛り（単位）に対応づける操作である。偏差値への変換を思い出してほしい。偏差値への変換では，標準化後数値 * 10 + 50 をして得られたデータを平均50，標準偏差10となるようにデータ（の目盛り）を調整している。z の目盛りから x への目盛りの変換では，x と z の相関を考慮して，$SD_x$ に $r_{xz}$ をかけて目盛りを調整していると考えることができる。

このように x の目盛りに変換された z の値は，相関関係に基づいて z から導かれた x の**予測値**といえる。この予測値を実際の x から引いた値が $x|z$ である。したがって，$x|z$は観測値と予測値の差，つまり，z では予測しきれなかった（相関関係では捉えきれなかった） x の残りの部分である。実際に**残差**（residual）と呼ばれる。

つまり，「z の影響を取り除く」は，「xのうち，相関関係に基づいてzから予測できる部分を取り除く」ことであるといえる。

極端な例を考えると分かりやすいかもしれない。$r_{xz}$= 0 のとき，式中の z に関する部分は 0 となり，$x|z = x - Mean_x$ となる。これは x を中心化しているだけであるから，x そのものである。つまり，全く相関がないと，z から x は予測できないため，残差も x のままである。

$r_{xz}$= 1 （あるいは -1） となる状況では， $z_{standardized}$ は $x_{standardized}$ （あるいは $-x_{standardized}$）に一致する（これは各自で確かめてほしい）。このとき，
$$
\begin{align}
z_{standardized} \times SD_x \times (\pm 1) + Mean_x &= \pm x_{standardized} \times SD_x \times (\pm 1) + Mean_x\\
&= x_{standardized} \times SD_x + Mean_x \\
&= x
\end{align}
$$

となり，第2項は x の標準化の逆の操作をしていることになる。したがって，$x|z$ = 0 になる。x, z が完全に相関するとき，z から x は完全に予測できるため，残差は 0となる。
