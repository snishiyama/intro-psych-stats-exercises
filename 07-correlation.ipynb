### 必要なPythonの文法知識：zip

for文を使って，複数のリストから対応する位置の要素を取り出すには，`zip`を使うと便利である。`zip` は以下のようにfor文のinの後に`zip(1つ目のリスト, 2つ目のリスト)` のように配置して利用する。

```python
for a, b in zip(listA, listB):
    何か処理
```

注目してほしいのは`for`の直後が `a, b` となっていることである。`zip`を使うときはこのようにすることで，`listA`と`listB`の対応する位置の値をそれぞれ`a`, `b`という変数として，forブロックの処理で利用できるようになる。以下のサンプルコードを実行して，具体的な挙動を確認してほしい


````python
list_a = [1, 2, 3, 4, 5]
list_b = [6, 7, 8, 9, 10]
for youso_a, youso_b in zip(list_a, list_b):
    print(youso_a + youso_b)
````
