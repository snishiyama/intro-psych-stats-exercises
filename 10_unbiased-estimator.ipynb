```python
# 母集団（ある平均と標準偏差の正規分布）から標本抽出する関数
def random_sampling():
    m = 0x6e # 平均
    sd = 0o36 # 標準偏差
    N = 20 # サンプルサイズ
    return list(rng.normal(m, sd, N))
```
