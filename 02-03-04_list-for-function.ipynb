# リスト

Pythonでリストを使うと，複数の値を一つの変数にまとめることができる。以下のように，複数の値をカンマ`,`で区切り，全体を `[]` で囲むとリストを作成できる。

````python
[1, 3, 5, 7, 9] # 奇数のリスト
````

リストに名前をつけておく，つまりリストの変数を宣言しておくこともできる。宣言しておくと，変数名を書くだけでリストにアクセスすることができる。繰り返し取り扱う場合には必須である。

以下の例でもここで宣言している`odd_list`を繰り返し使用する

````python
odd_list = [1, 3, 5, 7, 9] # odd は奇数のこと
````

統計学的な分析を行う場面において，データはリストのような形式で扱うことが多いので，リストの操作には慣れておかないといけない。

まず，リストの要素の数を調べるときは，`len(リストの変数名)` とする。

````python
len(odd_list) # リストの長さ（要素の数）。len は length の省略形
````

リストの個々の値にアクセスするためには，インデックス（index, 位置のこと）を利用する。リストの1番目の要素にアクセスしたかったら，`odd_list[0]`とする。2番目の場合は`odd_list[1]` とする

※※※**注意**※※※

日常的な数え方では位置を1から順に数えるが，Pythonでは0から数える。

````python
odd_list[0] # 日常的な数え方でいう1番目
````

````python
odd_list[1] # 日常的な数え方でいう2番目
````

````python
odd_list[2] # 日常的な数え方でいう3番目
````

用意されているインデックス以外を指定するとエラーになる

````python
odd_list[5] # 要素数は全部で5だが，インデックスの始まりは0のため，最後の要素のインデックスは 4 となり，5は存在しない
````

インデックスを使ってリストの中身を確認できるだけでなく，そのインデックスに新しい値を代入して上書きすることができる

````python
odd_list[0] = 21
odd_list
````

リストに要素を追加したければ，`リスト変数.append(追加したい要素)` とする。指定した要素はリストの最後に追加される

````python
odd_list.append(11)
odd_list
````

リストの要素を削除する方法もあるが，トラップも多いので，割愛する。

# for 文

統計学では，複数の値（=リストの各要素）に対して同じ処理を施すということがある。例えば，演習2のように，値すべてを2倍するということがあるかもしれない。その場合，（おそらく）演習2で行ったような，インデックスを使って各要素にアクセスする方法は，実用的ではない。今回くらいのリストの長さであればまだしも，実際のデータはもっと多いからである。

そういった繰り返し処理に便利なのが for 文である。

```python
for 任意の変数名 in リスト変数:
    行いたい処理
```

のように書き，リスト変数の各要素を順番に取り出して変数に格納し，処理を実施する。リストの要素を順番に取り出すので，要素の数の分だけ，処理が繰り返される。

次は，for 文のもっともシンプルな例で，リストの各要素を表示している。`print()`は`()`内に指定された値を明示的に表示するためのもの。

````python
for youso in [1, 2, 3]:
    print(youso) # print(...) は変数に入っている値を表示する
````

ここで次の2点に注意してほしい。

1. 1行目の最後がコロン`:` で終わっていることである。これを忘れるとエラーになる。エラーになったときは，まず，この`:`を忘れていないか確認してほしい。

2. 2行目が半角4文字分だけ字下げ（インデント）されている。Pythonではインデントにも文法上の意味があり，for文で繰り返したい処理の範囲を決めている。同じ幅のインデントの部分が，繰り返される処理になる。

以下のコードを実行してみると 2つ目のポイントについて理解できるはず。自分で処理を追加したり，インデントを変えてみたりすると，より理解が深まるだろう。

````python
for i in [1, 2, 3]: # 変数名はなんでもよい
    print(99)
    print(i) # ここまでの処理が繰り返される
print(10000) # これは1回だけ実行される
````

なお，`for i in ...` の `in` の後にはリストを直接書かなくても，リストと紐づいた変数名を指定することもできる。

````python
for odd in odd_list:
    print(odd)
````

上の`for`文は以下の処理と同様である。全体的な行数が少なくなっていることに注目。

````python
print(odd_list[0] * 2)
print(odd_list[1] * 2)
print(odd_list[2] * 2)
````

## for 文で合計を算出する

for文の活用場面は演習3のような新しいリストを作成することだけにとどまらない。演習3のコードを少し変更するだけで，リストに入っているすべての値の合計を算出することもできる。

````python
goukei = 0
atai_list = [1, 2, 3, 4, 5]
for atai in atai_list:
    goukei = goukei + atai # 取り出したリストの値をgoukeiに足して上書きする

goukei
````

# 関数

演習4では合計を算出するコードを2回繰り返し書いた。`for x in ...` の `...` 以外の部分の処理は共通である。コピペで対応するのもよいが，関数を用いることで，全体のコードを量を減らすことができる。

関数とは，一連の処理をまとめたもので，以下のように作成する

```python
def 関数名(引数1, 引数2, ...):
    まとめたい処理
    return 処理の結果
```

ここでもインデントを使ってどこまでが関数の定義かを区別できるようにしている。また，最後の`return`を使うことで，処理の結果を得ることができる。

例えば，底辺と高さのわかっている三角形の面積を求める関数は以下のように作成できる。

````python
def triangle(teihen, takasa):
    menseki = teihen * takasa / 2
    return menseki

triangle(5, 3)
````

一回定義すれば，名前だけで何度も呼び出すことができる

````python
triangle(29, 3)
````

引数にリストを指定するような関数を作成することもできる。例えば，演習2, 3で作成した，リストの中身を2倍するコードは以下のような関数にできる

````python
def nibai(atai_list):
    new_list = [] # 空のリスト
    for atai in atai_list:
        new_list.append(atai * 2)
    return new_list
````
