## ステップ3

ステップ2での実装に誤りがなければ，差の平均値と平均値差は等しいが，それらの標準誤差の結果は異なり差の平均値の標準誤差の方が小さいことがわかる。この理由について考察しよう。

### 平均値の場合

差の平均値は

$$
Mean_{A-B} = \frac{1}{n}\sum^{n}_{i = 1}(a_i - b_i) 
$$

平均値差は

$$
Mean_A-Mean_B = \frac{1}{n}\sum^{n}_{i = 1}a_i -\frac{1}{n}\sum^{n}_{i = 1}b_i
$$

どちらか一方の式を変形すると，以下の通り，もう一方の式に一致する（自分でやってみよう）



#### 式変形例

$$
\begin{array}{rcl}
\displaystyle
\frac{1}{n}\sum^{n}_{i = 1}(a_i - b_i) 
&=& \displaystyle\frac{1}{n}\big\{(a_1 - b_1) + (a_2 - b_2) + ⋯ + (a_n - b_n)\big\}\\
&=& \displaystyle\frac{1}{n}\big\{(a_1 + a_2 + ⋯ + a_n) - (b_1 + b_2 + \cdots + b_n)\big\} \\
&=& \displaystyle\frac{1}{n}\big(\sum^{n}_{i = 1}a_i -\sum^{n}_{i = 1}b_i\big) \\
&=& \displaystyle\frac{1}{n}\sum^{n}_{i = 1}a_i -\frac{1}{n}\sum^{n}_{i = 1}b_i
\end{array}
$$

### 標準誤差の場合

差の平均値（$Mean_{A-B}$）の標準誤差は

$$
SE_{Mean_{A-B}} = \sqrt{\frac{Var^{unbiased}_{A-B}}{n}}
$$

ただし，
$$
Var^{unbiased}_{A-B} = \frac{1}{n - 1}\sum^n_{i=1}\big\{(a_i-b_i) - Mean_{A - B}\big\}^2
$$

平均値差（$Mean_A - Mean_B$）の標準誤差は

$$
SE_{Mean_A - Mean_B} = \sqrt{\frac{Var^{unbiased}_A}{n}+\frac{Var^{unbiased}_B}{n}}
$$

ただし，
$$
Var^{unbiased}_A = \frac{1}{n - 1}\sum^n_{i=1}(a_i - Mean_A)^2\\
Var^{unbiased}_B = \frac{1}{n - 1}\sum^n_{i=1}(b_i - Mean_B)^2
$$

平均値差の標準誤差の式を変形すると

$$
\sqrt{\frac{Var^{unbiased}_A}{n}+\frac{Var^{unbiased}_B}{n}} = \sqrt{\frac{Var^{unbiased}_A+Var^{unbiased}_B}{n}}
$$

となり，$SE_{Mean_{A-B}}$と$SE_{Mean_A - Mean_B}$の式の違いは平方根の中にある分子だけである。以降では，$Var^{unbiased}_{A-B}$と$Var^{unbiased}_A+Var^{unbiased}_B$の関係についてのみ考える。

上で示した
$$
Mean_{A-B} = Mean_A - Mean_B
$$

という関係を利用すれば，$Var^{unbiased}_{A-B}$の式は以下のように変形できる

$$
\begin{array}{rcl}
Var^{unbiased}_{A-B} &=& \displaystyle\frac{1}{n - 1}\sum^n_{i=1}\big\{(a_i-b_i) - Mean_{A - B}\big\}^2\\
&=& \displaystyle\frac{1}{n - 1}\sum^n_{i=1}\big\{(a_i-b_i) - (Mean_A + Mean_B)\big\}^2\\
&=& \displaystyle\frac{1}{n - 1}\sum^n_{i=1}\big\{(a_i-Mean_A) - (b_i - Mean_B)\big\}^2\\
\end{array}
$$

さらにこの式を変形していけば，以下の式が得られる（自分でやってみよう）。

$$
Var^{unbiased}_{A-B} = Var^{unbiased}_A + Var^{unbiased}_B - 2 \times Cov_{A,B}
$$

$Cov_{A,B}$はAとBの共分散である。これが2つの標準誤差の差のもとである。具体的な差は以下の通りである。平方根記号のせいで，これ以上シンプルな形にはならない（たぶん）。

$$
\begin{array}{rcl}
SE_{Mean_{A-B}} - SE_{Mean_A - Mean_B} &=& \displaystyle\sqrt{\frac{Var^{unbiased}_A + Var^{unbiased}_B - 2 \times Cov_{A,B}}{n}} - \sqrt{\frac{Var^{unbiased}_A+Var^{unbiased}_B}{n}}\\
&=& \displaystyle\frac{1}{\sqrt{n}}\big(\sqrt{Var^{unbiased}_A + Var^{unbiased}_B - 2 \times Cov_{A,B}} - \sqrt{Var^{unbiased}_A+Var^{unbiased}_B}\big)\\
\end{array}
$$
