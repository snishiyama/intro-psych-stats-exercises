## 計算順序

```python
1 + 2 + 3
```

手計算と同様に，プログラミング言語の演算子にも計算の優先順位がある。担当教員の知る限り，手計算と同じ。

つまり，累乗 > 掛け算・割り算 > 足し算・引き算 の順

```python
1 + 2 * 3
```

```python
4 * 6 / 2 ** 3
```

手計算と同様に，括弧`()`を使うことでカッコ内の計算を優先させることができる。

```python
(1 + 2) * 3
```

ただし，手計算の場合とは異なり，使う括弧の種類は`()` のみで，`{}`や`[]`は四則演算には用いない。これらの記号はpythonでは特別な意味がある。

```python
(4 * (6 / 2)) ** 3
```

# 変数の宣言

プログラミング言語では，任意の値に名前をつけておくことができる。名前で呼び出すことでその値を繰り返し使うことができる

そういった名前をつけておいた値のことを**変数**と呼ぶ。値に名前をつけることを変数の宣言・定義などと言う

`名前 = 値` で変数の宣言を行う。**変数名はなんでも構わない**が，わかりやすい名前をつけるのがよい。

```python
hensu = 10
hensu / 2
```

一度宣言した変数は，同じファイルの他のコードブロックでも使用できる

```python
hensu - 0.9
```

計算結果を新しい変数名をつけることもできる

```python
kekka = hensu * 3
kekka
```

Pythonでは，一度宣言した変数の値をを上書きすることができる

```python
hensu = 25 # 10から25への上書き
hensu / 2
```

なお，このとき，上書き前の`hensu`を使ってすでに計算した`kekka`の値は更新されない。

```python
kekka
```

すでに宣言した変数を使った計算を行って，もとの変数を上書きすることもできる

```python
hensu = hensu + 2
hensu
```

数値の変数同士の演算も可能

```python
goukei = hensu + kekka
goukei
```

こういうこともできる

```python
goukei = goukei + hensu
goukei
```
